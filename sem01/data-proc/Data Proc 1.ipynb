## Часть 1. Интерполяция полиномами

### Пример 1

```python
N = 4
A = 0
B = 1
x_a = [1, 2, 3, 4, 5]
y_a = np.tan(x_a)
print(x_a)
plt.plot(x_a, y_a, 'v', markersize=8)
```

```python
poly_a = interp.lagrange(x_a, y_a)
plt.plot(x_a, y_a, 'v', markersize=8)
print(poly_a)
plt.plot(x_a, poly_a(x_a))
```

### Пример 2

```python
x_c = [1, 2, 3, 5, 6]
y_c = [9, -9, 9, 15, -15]
```

```python
poly_c = interp.lagrange(x_c, y_c)

print (poly_c)

x_d = np.arange(0, 10)

plt.plot(x_c, y_c, 'v')
plt.plot(x_d, poly_c(x_d))
```

```python
x_f = np.arange(-1.0, 1.25, 0.25)
y_f = 1/(1+25*x_f*x_f)
poly_f = interp.lagrange(x_f, y_f)
print (poly_f)
x_plot = x_d
plt.plot(x_f, y_f, 'v')
plt.plot(x_plot, poly_f(x_plot))
```

```python
plt.plot(x_f, y_f, 'v')
plt.plot(x_d, poly_f(x_d))
plt.plot(x_d, poly_c(x_d))
```
