```python
import pandas as pd
import numpy as np

#df = pd.read_csv("../maga/src/test/resources/iris.data")
#df = df.sample(frac=1).reset_index(drop=True)
#df.to_csv("../maga/src/test/resources/iris_shuffle.data", index=False)
df = pd.read_csv("../maga/src/test/resources/iris_shuffle.data")
df
```

#### 1. Get average sepal width

```python
df['sepalWidth'].mean()
```

#### 2. Get average petal square - petal width multiplied on petal length

```python
def petalSquare(df):
    return df['petalWidth'] * df['petalLength']

def averagePetalSquare(df):
    return petalSquare(df).mean()

averagePetalSquare(df)
```

#### 3. Get average petal square for flowers with sepal width > 4

```python
averagePetalSquare(df[df['sepalWidth'] > 4])
```

#### 4. Get flowers grouped by Petal size (Petal.SMALL, etc.)

```python
df['petalSquare'] = petalSquare(df)
df
```

```python
def classifyByPetal(x):
    if x < 2.0:
        return 'small'
    elif x < 5.0:
        return 'medium'
    return 'large'

df['class'] = df['petalSquare'].apply(classifyByPetal)
df
```

```python
grouped = df.groupby('class')
```

```python
small = grouped.get_group('small')
print(small.shape[0])
small
```

```python
medium = grouped.get_group('medium')
print(medium.shape[0])
medium
```

```python
large = grouped.get_group('large')
print(large.shape[0])
large
```
